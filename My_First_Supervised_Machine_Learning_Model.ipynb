```
import numpy

x = 3 * numpy.random.rand(500, 1)
y = 3 * x + numpy.random.randn(500, 1)

import matplotlib.pyplot as pyplot

pyplot.figure(figsize=(8,6))
pyplot.scatter(x,y)
pyplot.title("Random Dataset")
pyplot.xlabel("X")
pyplot.ylabel("Y")
pyplot.show()

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=3/4, random_state=0)
x_train = x_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
x_test = x_test.reshape(-1 ,1)
y_test = y_test.reshape(-1, 1)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model = model.fit(x_train, y_train)
prediction = model.predict(x_test)
score = model.score(x_test, y_test)
print(score)

pyplot.scatter(x_train, y_train, color = "green", label = "Training Dataset")
pyplot.scatter(x_test, y_test, color = "orange", label = "Testing Dataset")
pyplot.legend()
pyplot.plot(x_train, model.predict(x_train), color = "red")
pyplot.title("Model Results")
pyplot.xlabel("X")
pyplot.ylabel("Y")
pyplot.show()
```
